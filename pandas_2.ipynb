## Filtering Data

```python
import pandas as pd
```

```python
df=pd.read_csv('pokemon_data.csv')
```

```python
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')] #use '&' instead of 'and' in pandas
```

```python
new_df=df.loc[(df['Type 1']=='Fire') & (df['Type 2']=='Flying') & (df['HP']>60)]
```

```python
new_df #the index values don't start from 1
```

```python
df.loc[df['Name'].str.contains('Mega')] #.str because we have to first convert 'Name' to type str
```

```python
import re
```

```python
df.loc[df['Name'].str.contains('^pi[a+z]*',flags=re.I,regex=True)] #re.I is to ignore the case sensitivity
```

## Resetting Index values

```python
new_df=new_df.reset_index() #this however retains the values of the old indices to a new column
```
