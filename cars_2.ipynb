## Control Statements 

```python
import pandas as pd
```

```python
df=pd.read_csv('Toyota.csv',index_col=0,na_values=['??','????'])
```

```python
df.insert(1,'Price_List','') #insert new column at position 2 with blank values
```

```python
df.head(5)
```

```python
for i in range(len(df['Price'])):
    if df['Price'][i]<=8750:
        df['Price_List'][i]='Low'
    elif df['Price'][i]>11750:
        df['Price_List'][i]='High'
    else:
        df['Price_List'][i]='Medium'
```

```python
df['Price_List'].value_counts()
```

## Functions

```python
df.insert(3,'Age_Converted',0)
```

```python
def convert(val):
    converted_value=val/12
    return converted_value

df['Age_Converted']=convert(df['Age'])
df['Age_Converted']=round(df['Age_Converted'],1)
```

```python
df.head()
```

```python
df.insert(5,'KM_Per_Month',0)
```

```python
def convert(val1,val2):
    converted_value=val1/12
    ratio=val2/val1
    return [converted_value,ratio]

df['Age_Converted'],df['KM_Per_Month']=convert(df['Age'],df['KM'])
```
