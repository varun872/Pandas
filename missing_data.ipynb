```python
df1['Age'].fillna(df1['Age'].mean(),inplace=True)
```

```python
df1['KM'].fillna(df1['KM'].median(),inplace=True)
```

```python
df1['HP'].fillna(df1['HP'].mean(),inplace=True)
```

```python
# For categorical data fill NaN values with the most occuring value
df1['FuelType'].value_counts()
```

```python
df1['FuelType'].fillna(df1['FuelType']=='Petrol',inplace=True)
```

```python
df1['MetColor'].value_counts()
```

```python
df1['MetColor'].fillna(df1['MetColor'].mode()[0],inplace=True)
```
