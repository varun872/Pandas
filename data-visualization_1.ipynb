### Data Visualization: Allows us to quickly interpret the data and adjust the values according

#### Why to visualize data?
- To check for patterns
- Identifiy extreme values that can be anomalies

#### Matplotlib
- Matplotlib is a 2D plotting library which produces good quality figures

#### Scatter Plot
- Values are scattered btw 2 different axis
- They are used to get the correlation of 2 variables

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Toyota.csv',index_col=0,na_values=['??','????'])
```

```python
df.dropna(axis=0,inplace=True)
```

```python
plt.scatter(df['Age'],df['Price'],c='red')
plt.title('Plot of Age vs Price')
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

### Histogram
- Graphical representation of data using bars of different heights

```python
plt.hist(df['KM']) #histogram with default arguements (since y-axis value has not been given)
```

```python
plt.hist(df['KM'],color='green',edgecolor='grey',bins=5)
plt.title('Histogram of Kilometer')
plt.xlabel('KM')
plt.ylabel('Default')
plt.show()
```

### Bar Graph
- Plot that represents categorical data
- Similar to histogram but there is a separation btw the bars
- Bar Graph makes it easy to compare values of data btw different groups

```python
counts=[569,422,100]
fueltype=('Petrol','Diesel','CNG')
index=np.arange(len(fueltype))
```

```python
plt.bar(index,counts,color=['cyan','yellow','red'])
plt.title('Bar plots of fueltypes')
plt.xlabel('Index')
plt.ylabel('Fueltypes')
plt.xticks(index,fueltype)
plt.show()
```
