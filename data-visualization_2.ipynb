### Seaborn: Provides a high level interface for drawing informative graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Toyota.csv',index_col=0,na_values=['??','????'])
```

```python
df.dropna(axis=0,inplace=True) #removing NaN values in all rows
```

```python
# Scatter Plot
sns.set_theme(style='darkgrid')
sns.scatterplot(x=df['Age'],y=df['Price'])
```

```python
sns.lmplot(x='Age',y='Price',data=df,hue='FuelType',fit_reg=False,legend=True,palette='Set1')
```

```python
# Histogram with default kernel density estimate
sns.displot(df['Age'])
```

```python
sns.displot(df['Age'],kde=False,bins=5)
```

```python
# Bar Plot
sns.countplot(x='FuelType',data=df)
```

```python
# Grouped Bar Plot
sns.countplot(x='FuelType',data=df,hue='Automatic')
```

```python
# Boxplot
sns.boxplot(y=df['Price'])
```

```python
sns.boxplot(y='Price',data=df,x='FuelType')
```

```python
# Grouped Boxplot
sns.boxplot(y='Price',x='FuelType',hue='Automatic',data=df)
```

```python
#  boxplot and histogram on same graph
f,(ax_box,ax_hist)=plt.subplots(2,gridspec_kw={"height_ratios":(.15,.85)})
```

```python
# Pairwise plots
sns.pairplot(df,kind='scatter',hue='FuelType')
plt.show()
```
